# Anaconda 使用教程
---

## 1. 管理 conda：
---

### 1.1 检查 conda 已经被安装

```
conda --version
```

### 1.2 升级当前版本的 conda

```
conda update conda
```

### 1.3 conda 下载慢问题解决

```
conda config --add channels 'https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free/'
conda config --set show_channel_urls yes
```

## 2. 管理环境
---

### 2.1 创建并激活一个环境

```
# 基于 python3.6 创建一个名为test_py3 的环境
conda create --name test_py3 python=3.6 

# 基于 python2.7 创建一个名为test_py2 的环境
conda create --name test_py2 python=2.7

# 确定环境添加成功
conda info -e

# 激活 test 环境
source activate test_py2 

# 检查新的环境中的python版本
python --version

# 切换到python3
activate test_py3

# 注销该环境
source deactivate
```

### 2.2 列出所有的环境

```
conda info --envs
```

### 2.3 确认当前环境

```
conda info -envis
```

### 2.4 复制一个环境

```
conda create -n flowers --clone test_py3 
```

### 2.5 删除一个环境

```
conda remove -n flowers --all

# 为了确定这个名为flowers的环境已经被移除，输入以下命令：
conda info -e
```

### 2.6 学习更多关于环境的知识

如果你想学习更多关于conda的命令，就在该命令后边跟上 `-h`

```
conda remove -h
```

## 3. 管理 python
---

### 3.1 检查 python 版本

首先让我们检查那个版本的python可以被安装：

```
conda search --full --name python
```

## 4. 包管理
---

### 4.1 查看该环境中包和其版本的列表

使用这条命令来查看哪个版本的python或其他程序安装在了该环境中，或者确保某些包已经被安装了或被删除了。在你的终端窗口中输入：

```
conda list
```

### 4.2 查找一个包

```
conda search beautifulsoup4
```

### 4.3 安装一个包

```
conda install --name  python3.6 beautifulsoup4
```

**提示：** 你必须告诉conda你要安装环境的名字（-n  python3.6）否则它将会被安装到当前环境中。
