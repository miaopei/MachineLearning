```python
from fishery_model import Fishery_Model
import tensorflow as tf
import numpy as np
from data_utils import load_data
import os
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
%load_ext autoreload
%autoreload 2
```

```python
X_tr, Y_tr, X_val, Y_val, X_te, Y_te = load_data('data')
```

```python
tf.reset_default_graph()
model = Fishery_Model()
sess = tf.Session()
sess.run(tf.global_variables_initializer())
model.load_weights(os.path.join("vgg16", "vgg16_weights.npz"), sess)
```

```python
indicies = np.arange(Y_val.shape[0])
np.random.shuffle(indicies)
indicies = indicies[:4]
```

```python
print(Y_val[indicies])
for i in indicies:
    img = plt.imread(X_val[i])
    plt.imshow(img)
    plt.title(str(Y_val[i]) + ' - ' + X_val[i])
    plt.show()
```

```python
model.fit(
    sess,
    3,
    8,
    (X_tr, Y_tr),
    (X_val, Y_val),
    1.0
)
```
