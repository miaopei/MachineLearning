```python
from vgg16 import VGG16
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
```

```python
weights_file = 'vgg16_weights_tf_dim_ordering_tf_kernels.h5'

model = VGG16()
model.build((224, 224, 3))
model.load_weights(weights_file)

model.summary()
```

```python
from PIL import Image
from imagenet_classes import class_names

img = Image.open('laska.png').convert('RGB').resize((224,224))
x = np.asarray(img)
plt.imshow(x)
x = np.array([x])
```

```python
probs = model.predict(x)[0]
```

```python
p = np.argmax(probs)
```

```python
print(class_names[p])
```

```python
model2 = VGG16()
model2.build((224,224,3), False)
model2.summary()
```

```python
weights_file = 'vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5'

model2.load_weights(weights_file)
```

```python
from PIL import Image
from imagenet_classes import class_names

img = Image.open('cat.jpg').convert('RGB').resize((224,224))
x = np.asarray(img)
plt.imshow(x)
x = np.array([x])
```

```python
p = np.argmax(probs)
```
